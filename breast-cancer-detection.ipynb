## Correlation matrix

```python
X = df.drop('diagnosis',axis=1)
output = df['diagnosis']

corr_matrix = df.corr().abs()

# select the upper triangle
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

high_corr_features = [column for column in upper.columns if any(upper[column] > 0.9)]

# Create a dictionary of correlated feature groups
correlated_groups = {}
for feature in high_corr_features:
    correlated_with = [col for col in X.columns if corr_matrix.loc[feature, col] > 0.9 and col != feature]
    if correlated_with:
        group = tuple(sorted([feature] + correlated_with))
        correlated_groups[group] = corr_matrix.loc[list(group), list(group)]

print(f"Found {len(high_corr_features)} highly correlated features:")
for group in correlated_groups:
    print(f"  - Correlated group: {', '.join(group)}")
        
```

<h4>Optuna</h4>
<p>Optuna is an open-source hyperparameter optimization framework designed to automate the search for the best hyperparameters for machine learning algorithms. It provides a simple and flexible API to minimize the effort required to perform hyperparameter tuning, supporting various optimization algorithms. <h4>AdaBoostClassifier</h4>
    <p><code>AdaBoostClassifier</code> is a part of the sklearn.ensemble module, which implements the Adaptive Boosting algorithm for classification tasks. AdaBoost works by combining multiple weak classifiers to create a strong classifier. It iteratively adjusts the weights of incorrectly classified instances, allowing subsequent classifiers to focus more on difficult cases.</p>

```python
classifiers = {
    "Logistic Regression": LogisticRegression( max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(max_depth=15),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, validation_fraction=0.1, n_iter_no_change=10),
    "AdaBoost": AdaBoostClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}
train_accuracies = {}
test_accuracies = {}

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_train_pred = clf.predict(X_train)
    y_test_pred = clf.predict(X_test)
    train_accuracies[name] = accuracy_score(y_train, y_train_pred)
    test_accuracies[name] = accuracy_score(y_test, y_test_pred)


results = []
for name in classifiers.keys():
    results.append({
        'Model': name,
        'Train Accuracy': train_accuracies[name],
        'Test Accuracy': test_accuracies[name]
    })
pd.DataFrame(results)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve,average_precision_score
```

```python
model = AdaBoostClassifier(algorithm='SAMME', learning_rate=1.5, n_estimators=100)
model.fit(X_train,y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Get probabilities for ROC curve
y_test_proba = model.predict_proba(X_test)[:, 1]

print(f'Training accuracy: {accuracy_score(y_train, y_train_pred):.4f}')
print(f'Testing accuracy: {accuracy_score(y_test, y_test_pred):.4f}')

plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

fpr, tpr, _ = roc_curve(y_test, y_test_proba)
auc = roc_auc_score(y_test, y_test_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f'AUC Score: {auc:.4f}')

precision, recall, _ = precision_recall_curve(y_test, y_test_proba)
average_precision = average_precision_score(y_test, y_test_proba)

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'AP = {average_precision:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print("Classification report: {}".format(classification_report(y_test,y_test_pred)))
```
